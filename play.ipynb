Variables
1. *taxonomy*: microbial taxonomic classification, (a) domain, (b) phylum, (c) class, (d) order 
2. *abundance*: numerical 
3. *lake*: categorical (one of: Inre, Villa)
4. *edge*: categorical (one of: Center, Edge)
5. *depth*: categorical (one of: 2, 6, 10, 14, 16)
6. *sample type*: categorical (one of: Field, Incubation)
7. *sampling day*: only if sample type is "Incubation"; categorical (one of: 3, 10)
8. *temp*: categorical; 0 if sample type is "Field"; one of 10, 15, 20, 25 if sample type is "Incubation"
9. *sampling timepoint*: categorical (one of: Field, Initial, Subsequent)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('microbe_data.csv')
df = df.iloc[:, 1:] # drop first col 
df.head()
```

```python
print(f'# obs = {len(df)}')
```

```python
# var 1: taxonomy 
print(f"# unique: {len(df['taxonomy'].unique())}")
print(f"Undefined = {(df['taxonomy'] == 'Undefined').sum()}")
```

```python
# var 2: abundance 
print(f"0 = {(df['Abundance'] == 0).sum()}")
print(df['Abundance'].describe())
plt.hist(df[df['Abundance'] > 0]['Abundance'], bins=30, edgecolor='black')
plt.xlabel('Abundance')
plt.ylabel('Frequency')
plt.title('Non-zero Abundance (28848 - 16553 = 12295)')
plt.show()
```

```python
# var 3: lake 
print(f"# unique: {df['Lake'].unique()}")
print(f"NA = {df['Lake'].isna().sum()}")
```

```python
# var 4: edge 
print(f"# unique: {df['Edge'].unique()}")
print(f"NA = {df['Edge'].isna().sum()}")
```

```python
# var 5: depth 
print(f"# unique: {df['Depth'].unique()}")
print(f"NA = {df['Depth'].isna().sum()}")
```

```python
# var 6: sample type 
print(f"# unique: {df['SampleType'].unique()}")
print(f"NA = {df['SampleType'].isna().sum()}")
```
